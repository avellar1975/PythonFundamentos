```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

plt.figure(figsize=(12,10))
ax6 = sns.barplot(x="fuelType", y="price", data=df, estimator=np.mean, hue='gearbox')
plt.ylabel('Preço médio', fontsize=17)
plt.xlabel('Tipos de combustível', fontsize=17)
plt.title('Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig = ax6.get_figure()
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

plt.figure(figsize=(28,10))
ax7 = sns.barplot(x="brand", y="powerPS", data=df, estimator=np.mean, hue='gearbox')
plt.ylabel('Potência média', fontsize=17)
plt.xlabel('Tipos de veículos', fontsize=17)
plt.title('Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio', fontsize=20)
ax7.set_xticklabels(ax7.get_xticklabels(),rotation = 90)
plt.show()
```

```python
# Salvando o plot
fig = ax7.get_figure()
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
