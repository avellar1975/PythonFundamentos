```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
```

```python
df.columns
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
plt.figure(figsize=(12,8))
ax4 = sns.countplot(y='brand', data=df, orient='h')
plt.ylabel('Marcas', fontsize=17)
plt.xlabel('Número de veículos', fontsize=17)
plt.title('Veículos por marca', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig = ax4.get_figure()
fig.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
from numpy import mean
```

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

plt.figure(figsize=(28,10))
ax5 = sns.barplot(x="brand", y="price", data=df, estimator=mean, hue='gearbox')
plt.ylabel('Preço médio', fontsize=17)
plt.xlabel('Tipos de veículos', fontsize=17)
plt.title('Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio', fontsize=20)
ax5.set_xticklabels(ax5.get_xticklabels(),rotation = 90)
plt.show()
```

```python
# Salvando o plot
fig = ax5.get_figure()
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
