```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

### Breve análise do dataset

```python
df.columns
```

```python
df.info()
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

plt.figure(figsize=(12,8))
ax = sns.histplot(x=df['yearOfRegistration'], kde=True, bins=40, color='g')
plt.ylabel('Veículos')
plt.xlabel('Ano de Registro')
plt.title('Distribuição de Veículos por ano de registro', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig = ax.get_figure()
fig.savefig("plots/Analise1/vehicle-distribution.png")
# Para abrir a imagem <img src="plots/Analise1/vehicle-distribution.png">
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
# Faixa de preços ['price'] por tipo de veículo ['vehicleType']
plt.figure(figsize=(12,8))
ax2 = sns.boxplot(data=df, x='vehicleType', y='price')
plt.ylabel('Faixa de Preços')
plt.xlabel('Tipo de Veículo')
plt.title('Variação da faixa de preço pelo tipo de veículo', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig = ax2.get_figure()
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
# Para abrir a imagem <img src="plots/Analise1/price-vehicleType-boxplot.png">
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
plt.figure(figsize=(12,8))
ax3 = sns.countplot(x='vehicleType', data=df)
plt.ylabel('Total de veículos')
plt.xlabel('Tipo de Veículo')
plt.title('Total de veículos à venda por tipo', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig = ax3.get_figure()
fig.savefig("plots/Analise1/count-vehicleType.png")
```
